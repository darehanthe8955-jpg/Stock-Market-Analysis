```python
import yfinance as yf

df = yf.download("^NSEI", period="1y", interval="1d")
df.info()
df.describe()
df
```

```python
df.drop_duplicates()
df.dropna().copy()
df.reset_index(inplace =True)
df
```

```python
import pandas as pd
```

```python
df= df.sort_index()
df
```

```python
df["MA-50"] = df["Close"].rolling(window= 50).mean()
df["MA-200"] = df["Close"].rolling(window= 200).mean()

df["daily return"] = df["Close"].pct_change()
df["Volatility"] = df["daily return"].rolling(window = 20).std()
```

```python
import matplotlib.pyplot as plt

df["target"] = (df["Close"].shift(-1) > df["Close"]).astype(int)
```

```python
df = df.dropna().copy()
df.reset_index(drop=True, inplace=True)
df
```

```python
features = ['MA-50', 'MA-200', 'Volatility']

X = df[features]
y = df['target']
```

```python
split = int(len(df) * 0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=42
)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
df_test = df.iloc[split:].copy()
df_test.reset_index(drop=True, inplace=True)

df_test['Prediction'] = y_pred
df
```

```python
df.columns = df.columns.get_level_values(0)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df_test['target'].values, label='Actual')
plt.plot(df_test['Prediction'].values, label='Predicted', linestyle='--')

plt.legend()
plt.title("Random Forest Prediction")
plt.show()
```

```python
import pandas as pd

importance = pd.Series(model.feature_importances_, index=X.columns)
print(importance.sort_values(ascending=False))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
